```python
# Your first function should be defined in this cell. Each subsequent function will entered into a new cell. 

import pandas as pd 

def get_grades():
    '''retrieves files and returns dataframe of CSV'''
    #FILE_PATH = FILE_PATH = "C:\CIS5357\Assignments"
    #GRADES_FILE = FILE_PATH + '\Grades.csv'
    
    FILE_PATH = FILE_PATH = '/users/cis5357/assignments/'
    GRADES_FILE = FILE_PATH + 'Grades.csv'
    
    df_grades = pd.read_csv(GRADES_FILE)
   
    return df_grades 
```

```python
def get_dept_chair_input(df_grades):
    '''Gets user input and slices df for desired values, displays values in console to desired instructions'''
    want_continue = 'Y'
    count_of_grades = (df_grades.sum(axis = 0)) 
    count_of_grades = count_of_grades.rename("count")
    count_of_grades = count_of_grades.iloc[1:]

    avg_of_grades = df_grades.iloc[ : , 1: ].mean(axis = 0) 
    avg_of_grades = avg_of_grades.rename("average")
    
    
    
    med_of_grades = df_grades.iloc[ : , 1: ].median(axis = 0)
    med_of_grades = med_of_grades.rename("median")
    
    min_of_grades = df_grades.iloc[ : , 1: ].min(axis = 0)
    min_of_grades = min_of_grades.rename("lowest number of times")
    
    max_of_grades = df_grades.iloc[ : , 1: ].min(axis = 0)
    max_of_grades = max_of_grades.rename("Highest number of times")
    
    

    while want_continue.upper() in ('Y','Yes'):
        
        user_input = input("which letter grade would you like information on(eg. A,B,C)?: \t").upper()
        
        print(f'\n\nThe total count of times {user_input} letter grade was assigned : {count_of_grades[user_input]:>25}')
        print(f'The average count of the number of times {user_input} letter grade was assigned : {avg_of_grades[user_input]:>11.1f}') 
        print(f'The median number of times {user_input} letter grade was assigned : {med_of_grades[user_input]:>25.1f}') 
        print(f'The lowest number of times {user_input} letter grade was assigned : {min_of_grades[user_input]:>23}') 
        print(f'The highest number of times {user_input} letter grade was assigned : {max_of_grades[user_input]:>22}') 
        want_continue = input("\nDo you want to enter a different letter grade (Y or N)?:  ").strip()
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

def display_charts(df_grades):
    '''invokes dataframe established in first retrieval, then invokes two functions for side by side and vertical stack'''
    side_by_side(df_grades)
    
    vertical_stack_1_fig(df_grades)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

def side_by_side(df_grades):
    '''invokes dataframe and displays visualizations of letergrades in side by side format'''
    df_grades_a = df_grades['A']    
    df_grades_b = df_grades['B']
    df_grades_c = df_grades['C']
    df_grades_sem = df_grades['Semester']
    
    plt.style.use('seaborn')
    
    fig1 = plt.figure(figsize= (50,10), dpi = 120)  #complete figure size that contains all graphs
    fig1.suptitle('Comparison of Letter Grades Over Last 20 Semesters', fontsize = 40)
    
    graph_a = fig1.add_subplot(1,3,1)  #defines size of ax1 inside fig1
    graph_a.grid(True)
    graph_a.set_title('Letter Grades A', fontsize = 30)
    graph_a.set_xlabel('Semesters', fontsize= 25)
    graph_a.set_ylabel('Times Grade A was given', fontsize= 25)
    graph_a.tick_params(axis = 'y', labelsize=20, direction ='out', length=10)
    graph_a.tick_params(axis = 'x', direction='out', length=10, rotation=45, labelsize=20)
    graph_a.plot(df_grades_sem, df_grades_a, color='red')
    
    graph_b = fig1.add_subplot(1,3,2)  #defines size of ax1 inside fig1
    graph_b.grid(True)
    graph_b.set_title('Letter Grades B', fontsize = 30)
    graph_b.set_xlabel('Semesters', fontsize= 25)
    graph_b.set_ylabel('Times Grade B was given', fontsize= 25)
    graph_b.tick_params(axis = 'y', labelsize=20, direction ='out', length=10)
    graph_b.tick_params(axis = 'x', direction='out', length=10, rotation=45, labelsize=20)
    graph_b.plot(df_grades_sem, df_grades_b, color='green')
    
    graph_c = fig1.add_subplot(1,3,3)  #defines size of ax1 inside fig1
    graph_c.grid(True)
    graph_c.set_title('Letter Grades C', fontsize = 30)
    graph_c.set_xlabel('Semesters', fontsize= 25)
    graph_c.set_ylabel('Times Grade C was given', fontsize= 25)
    graph_c.tick_params(axis = 'y', labelsize=20, direction ='out', length=10)
    graph_c.tick_params(axis = 'x', direction='out', length=10, rotation=45, labelsize=20)
    graph_c.plot(df_grades_sem, df_grades_c, color='blue')
    
    plt.show();
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)

def vertical_stack_1_fig(df_grades):
    '''invokes dataframe and visualizes data in vertical stack format '''
    df_grades_a = df_grades['A']    
    df_grades_b = df_grades['B']
    df_grades_c = df_grades['C']
    df_grades_sem = df_grades['Semester']
    
    fig2 = plt.figure()
    
    fig2, (graph_a, graph_b, graph_c) = plt.subplots(3,1, figsize=(20,30))
    
    graph_a.grid(True)
    graph_a.set_title('Letter Grades A', fontsize = 30)
    graph_a.set_xlabel('Semesters', fontsize= 25)
    graph_a.set_ylabel('Times Grade A was given', fontsize= 25)
    graph_a.plot(df_grades_sem, df_grades_a, color='red')
    
    graph_b.grid(True)
    graph_b.set_title('Letter Grades B', fontsize = 30)
    graph_b.set_xlabel('Semesters', fontsize= 25)
    graph_b.set_ylabel('Times Grade B was given', fontsize= 25)
    graph_b.plot(df_grades_sem, df_grades_b, color='green')
    
    graph_c.grid(True)
    graph_c.set_title('Letter Grades C', fontsize = 30)
    graph_c.set_xlabel('Semesters', fontsize= 25)
    graph_c.set_ylabel('Times Grade C was given', fontsize= 25)
    graph_c.plot(df_grades_sem, df_grades_c, color='blue')
    
    fig2.tight_layout(pad=5)
    
    plt.show();
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
def stand_alone(df_grades):
    '''invokes dataframe and visualizes data in one graph in more comparable visual'''
    df_grades_a = df_grades['A']   
    df_grades_b = df_grades['B']
    df_grades_c = df_grades['C']
    df_grades_sem = df_grades['Semester']
    
    fig3 = plt.figure(figsize= (30,10), dpi = 120)
    fig3.suptitle('Overlay of all Letter Grades(A,B, C)', fontsize=26)
    graph_abc = fig3.add_subplot(1,1,1)
    graph_abc.set_title('Grade Distributions Over Last 20 Semesters', fontsize=30)
    graph_abc.set_xlabel('Semesters', fontsize = 24)
    graph_abc.set_ylabel('Times Grade Was Given', fontsize = 24)
    
    graph_abc.tick_params(axis='y', labelsize=15, direction ='out', length=4)
    graph_abc.tick_params(axis='x', direction='out', length=4, which= 'both', labelsize=15)
    
    graph_abc.xaxis.set_minor_locator(AutoMinorLocator(5))
    graph_abc.minorticks_on
    
    max_yvalue = max(df_grades_a.max(), df_grades_b.max(), df_grades_c.max())
    graph_abc.set_ylim(0,max_yvalue+3)
    
    x_axis = df_grades_sem
    graph_abc.set_xlim(0, len(df_grades_sem) +1)
    
    graph_abc.grid(True, which='both', linewidth=3, color='white')
    graph_abc.plot(x_axis, df_grades_a, label= 'Letter Grade A', color='red', marker='d', markersize=10)
    graph_abc.plot(x_axis, df_grades_b, label= 'Letter Grade B', color='green', marker='s', markersize=10, linestyle = '-.')
    graph_abc.plot(x_axis, df_grades_c, label= 'Letter Grade C', color='blue', marker= 'p', markersize=10, linestyle = 'dotted')
    graph_abc.legend(loc='upper right', fontsize = 15)
    
    plt.show();
```

```python
def main():
    '''drives logic for program'''
    df_grades = get_grades()
    get_dept_chair_input(df_grades)
    display_charts(df_grades)
    stand_alone(df_grades)
```

```python
main()
```
