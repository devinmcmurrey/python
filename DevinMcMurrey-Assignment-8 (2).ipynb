### Program Name : Display Brewery Information

### Program Objective: Consume JSON data retrieved via an Web API, manipulate the data and present a simple report in the python console. 

<font size = 1>
    

    
</font>
    

```python
# Your program code should begin in this cell, starting on line 3

def get_json():
    '''load json from .json file''' 
    import json
    import requests
    
    #input
    get_brewery_city = input(f"\nPlease enter the city for which you'd like to display brewery info:  ").strip()
    get_brewery_type = input(f"Please enter the type of breweries in {get_brewery_city} you wish to dispaly info for:  ").lower().strip()
    
    #process
    website_string = (f"https://api.openbrewerydb.org/breweries?by_city={get_brewery_city}&by_type={get_brewery_type}")
    website_response = requests.get(website_string)
    
    brewery_db = json.loads(website_response.text)
    print(brewery_db)#: used this and saw I had created a list of dictionaries with many keys and values that could be called upon from the key names


    #output
    return brewery_db
```

```python
def process_and_print(brewery_db):
    import pandas as pd
    ''' void function: takes list of dictionaries and iterates through, pulling values and creating console output'''
    # print out the table header 
    #input
    print_header()
    
    #iterate through dictionary and pull values 
    
    #process
    
    for brew_dict in brewery_db:  
        brewery_name = brew_dict['name']
        brewery_state = brew_dict['state']
        brewery_city = brew_dict['city']
        brewery_type = brew_dict['brewery_type']
        
        
        #now you can print the row
        #output
        print(f"{brewery_name:<50}{brewery_state:<15}{brewery_city:<15}{brewery_type:<12}")                         
                                
    brew_dict_df = pd.DataFrame(brewery_db)
    display(brew_dict_df)
```

```python
def print_header():
    '''void function creates header for console output'''
    print(f"\n{'Brewery Name':<50}{'State':<15}{'City':<15}{'Type':<12}")
    print(f"{'-'*45:<50}{'-'*11:<15}{'-'*12:<15}{'-'*9:<12}")
    
```

```python
def main():
    ''' invokes other functions and drives logic of program'''
    brewery_db = get_json()
    process_and_print(brewery_db)
    print('\nProgram ended successfully')
```

```python
main()
```
